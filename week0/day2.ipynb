# LangChain LCEL with Groq (Llama 3.3) Architecture

This notebook demonstrates how to set up a complete LLM pipeline using **LangChain Expression Language (LCEL)** and the **Groq API**. The goal is to generate responses from the Llama 3.3 model using a structured prompt engineering approach.

---

## 1. Environment Variables and Dependencies Setup
First, we import the required modules and libraries. This includes `dotenv` to securely load environment variables, alongside LangChain's core components (LLM interface, Prompts, and Output Parsers).

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# .env file se api keys load karne ke liye
load_dotenv()
```

* **`load_dotenv()`**: This function reads a `.env` file in your current directory and loads keys (like `GROQ_API_KEY`) into the system environment variables, keeping sensitive credentials secure and out of the source code.

---

## 2. LLM Configuration (Groq Model Setup)
In this step, we initialize Groq's **Llama 3.3 70B** model.

```python
# Configure the LLM
llm = ChatGroq(
    model="llama-3.3-70b-versatile", 
    temperature=0.7,
    api_key=os.getenv("GROQ_API_KEY")
)
```

* **`model="llama-3.3-70b-versatile"`**: We are using Llama 3.3's versatile 70B model, which is optimized for complex reasoning tasks and high-speed execution.
* **`temperature=0.7`**: This parameter controls the randomness and creativity of the model. A value of `0.7` produces balanced, natural, and precise responses.
* **`api_key`**: Fetched safely via `os.getenv()` to keep the API key protected.

---

## 3. Prompt Template Definition
Here, we define a structured prompt template that establishes the context and behavior (persona) for the AI model.

```python
# Set up the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI engineering assistant."),
    ("user", "{question}")
])
```

* **`system` (System Message)**: Instructs the model on its role (e.g., "You are a helpful AI engineering assistant") ensuring it answers within a professional engineering context.
* **`user` (User Message)**: Uses a dynamic placeholder (`{question}`) which will be filled with the actual user query at runtime.

---

## 4. LangChain Expression Language (LCEL) Chain Creation
Using LangChain's **Pipe Operator (`|`)**, we stitch components together into a sequential pipeline (chain).

* **Execution Flow**: `Input Data` $\rightarrow$ `Prompt Template` $\rightarrow$ `LLM (Groq)` $\rightarrow$ `StrOutputParser`

```python
# Construct the chain using the pipe operator (|)
print("Prompt: ", prompt)
chain = prompt | llm | StrOutputParser()
print("Chain: ", chain)
```

* **`StrOutputParser()`**: By default, chat models return a complex AI message metadata object. This parser extracts only the raw text response from that object and converts it into a clean string format.

---

## 5. Chain Execution and Response Generation
Finally, we invoke the chain by passing our input payload to execute the pipeline and test the output.

```python
# Call the model via the chain
response = chain.invoke({"question": "What is an AI agent and why does it matter?"})
print(response)
```

* **`chain.invoke()`**: This method takes a dictionary `{ "question": "..." }`, inserts the value into the prompt placeholder, streams it over to the Groq API, processes it through the string parser, and outputs the final result printed above.
