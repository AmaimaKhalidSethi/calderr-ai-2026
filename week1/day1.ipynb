# Part 1 — What Problem Are Large Language Models Solving?

---

# Before LLMs

To understand Large Language Models (LLMs), we first need to understand the problem they were created to solve.

Humans communicate using language.

Examples:

- English
- Urdu
- Arabic
- Chinese

For humans, language feels natural.

If someone says:

"The cat sat on the mat."

You instantly understand:

- cat = animal
- sat = action
- mat = object

A computer does not.

For a computer this sentence is merely:

"The"
"cat"
"sat"
"on"
"the"
"mat"

just symbols.

The biggest challenge in Artificial Intelligence has always been:

> How do we teach computers to understand human language?

This field is called:

Natural Language Processing (NLP)

---

# What is NLP?

NLP = Natural Language Processing

It is the branch of AI that deals with:

- Text
- Speech
- Human language

Examples:

- Google Translate
- ChatGPT
- Siri
- Alexa
- Grammarly

All belong to NLP.

---

# Early NLP Systems

Before Deep Learning existed, NLP relied on rules.

Example:

Rule:

IF sentence contains:

"Hello"

THEN answer:

"Hi"

Example:

User:
Hello

Bot:
Hi

This works.

But what about:

User:
Hey

User:
Hi

User:
Good Morning

Now we need more rules.

Thousands of rules.

Millions of rules.

Soon the system becomes impossible to maintain.

---

# Why Rule-Based Systems Failed

Suppose we create rules for:

"How are you?"

But then users write:

- How are u?
- How r u?
- How have you been?
- What's up?

All mean nearly the same thing.

A rule-based system struggles because language is flexible.

Humans constantly invent new ways to say the same thing.

Rules cannot scale.

---

# Machine Learning Arrives

Researchers realized:

Instead of teaching rules manually...

let the computer learn patterns from data.

Example:

Feed thousands of examples:

Input:
"I love this movie"

Label:
Positive

Input:
"This movie is terrible"

Label:
Negative

Eventually the model learns:

love → positive

terrible → negative

without explicit rules.

This was a major improvement.

---

# The Limitation of Traditional Machine Learning

Traditional ML still required humans to manually design features.

Example:

Sentence:

"I absolutely love this movie"

Engineers might create features:

- contains "love"
- contains "movie"
- sentence length

The model only sees these handcrafted features.

Humans are still doing much of the work.

---

# Deep Learning Changes Everything

Deep Learning introduced neural networks.

Instead of humans manually creating features:

the network learns features automatically.

This was revolutionary.

The model starts discovering:

- grammar
- sentiment
- relationships
- context

directly from data.

---

# What is a Neural Network?

A neural network is a mathematical system inspired loosely by the brain.

Input:

"I love cats"

↓

Many layers of computation

↓

Output:

Positive sentiment

The network gradually learns patterns by adjusting numbers called:

weights

Training = adjusting weights.

---

# The Next Big Problem

Even neural networks struggled with language.

Why?

Language depends heavily on context.

Example:

"I went to the bank."

What is bank?

Possibilities:

- River bank
- Financial bank

Need more context.

Now read:

"I went to the bank to withdraw money."

Meaning becomes clear.

Language understanding requires context.

Lots of context.

This became the central challenge.

---

# What is a Language Model?

A Language Model is a system that predicts the next word.

Example:

Input:

"The sky is"

Possible next word:

"blue"

Input:

"I drink"

Possible next words:

- water
- tea
- coffee

The model learns probabilities.

Not rules.

Probabilities.

---

# Why Predicting the Next Word Is Powerful

At first glance this seems trivial.

But consider:

Input:

"Pakistan is located in"

Next word:

"Asia"

To predict this correctly, the model must know geography.

Input:

"2 + 2 ="

Next token:

"4"

Now it must know arithmetic.

Input:

"Newton discovered"

Next token:

"gravity"

Now it must know science.

By learning next-word prediction on massive datasets, the model accidentally learns enormous amounts of world knowledge.

This is the key insight behind modern LLMs.

---

# What Makes an LLM Different?

Language Model:

Predicts next word.

Large Language Model:

Predicts next word using:

- billions of parameters
- massive datasets
- huge computational resources

Examples:

- GPT
- Llama
- Mixtral
- Claude

The word "Large" simply means:

very large neural network.

---

# A Useful Mental Model

Imagine a student who has read:

- books
- articles
- websites
- research papers

for years.

Then you ask a question.

The student responds based on patterns learned from all that reading.

An LLM works similarly.

It does not search a database sentence-by-sentence.

Instead it has compressed statistical knowledge into billions of learned parameters.

---

# The Core Idea of Modern AI

Everything we call:

- ChatGPT
- GPT-4
- Llama
- Mixtral
- Claude

is fundamentally trying to solve one problem:

> Given previous text, predict the most likely next token.

Everything else emerges from this capability.

Translation,
summarization,
coding,
reasoning,
chatting,
question answering,

all emerge from large-scale next-token prediction.

---

# Key Takeaways

1. Computers do not naturally understand language.

2. Rule-based NLP did not scale.

3. Machine Learning learned patterns from data.

4. Deep Learning learned features automatically.

5. Language understanding requires context.

6. Language Models predict the next token.

7. Large Language Models are huge neural networks trained on massive datasets.

8. Modern AI systems are fundamentally next-token prediction machines.


# Part 2 — Tokens and Embeddings

---

# Where We Left Off

In Part 1 we learned:

A Language Model is simply a system that predicts the next token.

Example:

Input:

"The sky is"

Prediction:

"blue"

But now a huge question appears:

How does a computer even read words?

A neural network only understands numbers.

Not text.

Not English.

Not Urdu.

Not emojis.

Not code.

Only numbers.

So before an LLM can predict anything, it must convert text into numbers.

This is where:

1. Tokens
2. Embeddings

come into the picture.

These are among the most important concepts in all of LLMs.

---

# Why Computers Cannot Read Words

Consider the word:

cat

Humans instantly understand it.

A computer does not.

If we send:

cat

into a neural network, the network has no idea what that means.

Neural networks perform mathematics.

Mathematics requires numbers.

Therefore:

Text → Numbers

must happen first.

---

# Naive Approach #1

Suppose we assign:

cat = 1
dog = 2
fish = 3

Now the network receives numbers.

Problem:

The model may think:

dog (2)

is somehow twice cat (1)

which is nonsense.

Words do not have numerical relationships like that.

---

# Naive Approach #2: One-Hot Encoding

Imagine vocabulary:

cat
dog
fish

Represent each word as:

cat  = [1,0,0]
dog  = [0,1,0]
fish = [0,0,1]

Now every word gets its own position.

Better.

But another problem appears.

---

# Why One-Hot Encoding Fails

Look at:

cat  = [1,0,0]
dog  = [0,1,0]

Distance between cat and dog:

Same as:

cat and airplane

Same as:

cat and quantum physics

The model cannot understand similarity.

To the computer:

every word is equally unrelated.

But humans know:

cat and dog are similar.

cat and airplane are not.

We need a smarter representation.

---

# Enter Embeddings

Embeddings are one of the greatest ideas in modern AI.

Instead of assigning fixed vectors manually:

the neural network learns vectors automatically.

Example:

cat

might become:

[0.21, -0.88, 0.52, 1.33, ...]

dog

might become:

[0.19, -0.84, 0.55, 1.29, ...]

Notice:

These vectors are similar.

Why?

Because cat and dog often appear in similar contexts.

---

# The Magic Principle

The embedding for a word is learned from how that word is used.

Think about these sentences:

The cat chased the mouse.

The dog chased the ball.

The cat slept.

The dog slept.

The model notices:

cat and dog appear in similar situations.

Therefore:

their embeddings become close together.

---

# Visualizing Embeddings

Imagine a giant 3D universe.

Words become points in space.

Example:

cat      ●
dog      ●

near each other.

Meanwhile:

airplane ●

far away.

Because airplane appears in very different contexts.

---

# Example Embedding Space

Animals:

cat
dog
lion
tiger

cluster together.

Countries:

Pakistan
India
China

cluster together.

Programming:

Python
Java
C++

cluster together.

The model automatically learns these relationships.

Nobody explicitly teaches them.

---

# A Famous Example

Researchers discovered:

embedding(King)
- embedding(Man)
+ embedding(Woman)

≈ embedding(Queen)

The model learned this relationship from text.

Nobody hardcoded it.

This shocked researchers.

It showed that neural networks were learning meaningful concepts.

---

# What is a Token?

Now we need another critical idea.

LLMs do not actually process words.

They process tokens.

---

# Common Misconception

# Another Example

Instead:

Tokenization

# Tokens Become IDs

# Token IDs Are Not Embeddings

Not meaning.

Not knowledge.

Not understanding.

Just a lookup key.

---

# Embedding Lookup

The model now performs:

↓

Embedding Table

↓

Vector

Example:

Now the neural network has something meaningful to process.

---

# Think of a Dictionary

Example:

# Why Embeddings Are So Powerful

Embeddings solve several problems:

1. Numerical representation

Words become numbers.

2. Similarity

Related words become close.

3. Generalization

Model can understand unseen combinations.

4. Knowledge storage

Semantic relationships emerge naturally.

---

# Real Embedding Dimensions

Modern LLMs do not use tiny vectors.

Examples:

Each token becomes a huge vector.

---

# What Does One Dimension Mean?

Nothing individually.

Example:

Nobody knows.

The meaning emerges collectively.

Knowledge is distributed across thousands of dimensions.

This is called:

---

# Analogy: Human Concepts

Suppose I ask:

Instead:

- furry

---

# The Embedding Layer

The first trainable layer in an LLM is usually:

Embedding Layer

Its job:

Convert token IDs into meaningful vectors.

Without embeddings:

Transformers cannot work.

GPT cannot work.

Llama cannot work.

Nothing works.

---

# Key Takeaways

1. Neural networks understand numbers, not text.

2. Text must be converted into numerical representations.

3. One-hot encoding fails because it cannot represent similarity.

4. Embeddings are learned vector representations.

5. Similar words get similar embeddings.

6. LLMs process tokens, not words.

7. Tokenization converts text into tokens.

8. Tokens become token IDs.

9. Token IDs are looked up in an embedding table.

10. Embeddings are the first meaningful representation inside an LLM.

# The Long-Term Memory Problem

Word 1

↓

Word 2

↓

Word 3

↓

...

↓

Word 100

# Human Analogy

Person 1:

# The Vanishing Gradient Problem

This became one of the biggest problems in deep learning.

During training:

The model learns using gradients.

These gradients travel backward through the network.

In long sequences:

gradients become extremely small.

Almost zero.

This is called:

Vanishing Gradient

---

# What Happens?

If gradients vanish:

The network cannot learn long-range dependencies.

It forgets earlier information.

Example:

"The capital of France is Paris."

When predicting:

Paris

the model may struggle to remember:

France

if it occurred far away.

---

# Another Example

Sentence:

To understand:

French

the model needs information from very early in the sequence.

RNNs often fail here.

---

# Why This Was a Huge Problem

Language depends heavily on long-distance relationships.

Examples:

- Pronouns
- References
- Context
- Stories
- Conversations

RNNs struggled as sequences became longer.

---

# Enter LSTM

Researchers realized:

We need better memory.

This led to:

LSTM

Long Short-Term Memory Network

Introduced in 1997.

One of the most influential ideas in deep learning.

---

# What Was the Goal?

RNN Goal:

Remember previous information.

LSTM Goal:

Remember important information for much longer.

Forget unimportant information.

---

# Human Brain Analogy

Suppose someone tells you:

"My birthday is tomorrow."

Important.

You remember it.

Suppose someone tells you:

"I drank water at 3:17 PM."

Probably not important.

You forget it.

LSTM tries to learn similar behavior.

---

# LSTM's Core Idea

Instead of one memory,

LSTM introduces controlled memory.

It learns:

- What to remember
- What to forget
- What to update

using special gates.

---

# Gates

An LSTM contains:

1. Forget Gate

2. Input Gate

3. Output Gate

Think of gates as valves.

They control information flow.

---

# Forget Gate

Question:

Should this information be kept?

or discarded?

Example:

Temporary details may be forgotten.

Important context may be preserved.

---

# Input Gate

Question:

Should new information enter memory?

If important:

store it.

If irrelevant:

ignore it.

---

# Output Gate

Question:

What information should be revealed right now?

Not all memory is needed immediately.

---

# Why LSTMs Were Revolutionary

Compared to RNNs:

- Better memory
- Better long-range learning
- Better translation
- Better text generation

For many years:

LSTMs dominated NLP.

---

# Part 7 — Inference, Token Generation, Temperature, Top-k, Top-p, and Context Windows
# Introduction

Up until now we have focused on how Large Language Models are built and trained. We learned about tokens, embeddings, transformers, attention, parameters, pretraining, and fine-tuning.

However, once training is complete, the model enters a completely different phase called Inference.

Inference is the phase in which the model uses the knowledge already stored inside its parameters to generate responses. Every time you use ChatGPT, Claude, Llama, or Mixtral, you are interacting with a model that is performing inference.

Understanding inference is extremely important because this is the phase you will actually work with during your internship when using Groq APIs.

# What is Inference?

Inference is the process of using a trained neural network to make predictions.

For Large Language Models, inference means:

| Predicting the next token based on all previously seen tokens.

Many beginners think an LLM works like a human:

Question
↓
Thinking
↓
Answer

In reality, an LLM works more like:

Question
↓
Predict Next Token
↓
Add Token
↓
Predict Next Token
↓
Add Token
↓
Repeat

The model continuously predicts one token at a time until a complete response is generated.

# How an LLM Generates a Response

Suppose the user asks:

What is machine learning?

The model does not generate the entire answer at once.

Instead, it generates the answer token by token.

The first prediction might be:

Machine

Now the input becomes:

What is machine learning? Machine

The model performs another prediction.

The next token might be:

learning

The input now becomes:

What is machine learning? Machine learning

The process continues:

is

then

a

then

field

then

of

and so on.

Every response produced by a modern LLM is generated using this iterative process.

* Why Next-Token Prediction Appears Intelligent

A common question is:

If the model is only predicting the next token, why does it seem intelligent?

The answer lies in the scale of training.

To correctly predict:

The capital of France is ______

the model must learn geography.

To correctly predict:

Newton discovered ______

the model must learn science.

To correctly predict:

Write Python code that sorts a list

the model must learn programming.

As models become better at next-token prediction, they automatically learn enormous amounts of knowledge and reasoning patterns.

One of the biggest discoveries in modern AI is that sophisticated behavior emerges from large-scale next-token prediction.

** Understanding Probabilities **

When generating a token, the model does not usually decide on a single answer immediately.

Instead, it calculates probabilities for many possible next tokens.

Consider the prompt:

The capital of France is

The model may internally assign probabilities such as:

Token	Probability
Paris	91%
Lyon	4%
Marseille	3%
London	1%
Berlin	1%

These probabilities represent the model's confidence about possible continuations.

The next token is selected from this probability distribution.

* Temperature

Temperature is one of the most important generation parameters.

It controls how much randomness is introduced during token selection.

A *lower temperature* makes the model more deterministic.

A *higher temperature* makes the model more creative and unpredictable.

Temperature = 0

Temperature 0 tells the model to behave as deterministically as possible.

The highest-probability token is almost always selected.

Using the previous example:

Token	Probability
Paris	91%
Lyon	4%
Marseille	3%
London	1%
Berlin	1%

The model will almost always choose:

Paris

This makes temperature 0 useful for:

Programming
Mathematics
Data extraction
Structured outputs
API generation

Consistency is prioritized over creativity.

* Temperature = 1

Temperature 1 is generally considered a balanced setting.

Probabilities remain close to their original values.

The model *remains reasonably accurate* while still producing natural and varied responses.

Most conversational AI systems operate around this range.

This setting is often suitable for:

Chatbots
Educational explanations
General question answering
Everyday conversations
* Temperature = 2

Temperature 2 significantly increases randomness.

The probability distribution becomes flatter, allowing less likely tokens to be selected more often.

This can produce:

More creativity
More diverse outputs
Unexpected ideas

However, it can also produce:

More mistakes
Less factual accuracy
More hallucinations

High temperatures are often useful for:

Story generation
Brainstorming
Creative writing
Idea exploration

but are generally not ideal for factual or technical tasks.

# Intuition Behind Temperature

Imagine a classroom where students are answering questions.

Temperature 0

Only the smartest student is allowed to answer.

Temperature 1

The smartest student usually answers, but others occasionally contribute.

Temperature 2

Everyone starts shouting possible answers.

Some answers are brilliant.

Some answers are completely wrong.

This is essentially what temperature controls.

# Top-k Sampling

Even after applying temperature, the model may still consider thousands of possible tokens.

Many of these options are extremely unlikely and undesirable.

Top-k sampling *limits the model to only the k most probable* tokens.

For example:

Top-k = 5

The model keeps only the five most likely candidates and discards everything else.

This reduces the chance of selecting bizarre or irrelevant tokens.

Example of Top-k

Suppose the model predicts:

Token	Probability
Paris	50%
Lyon	20%
London	15%
Berlin	10%
Madrid	3%
Apple	1%
Banana	1%

With:

Top-k = 5

Only the first five tokens remain.

Apple and Banana are removed entirely.

The next token must be selected from the remaining candidates.

# Top-p Sampling

Top-p sampling is usually more flexible than Top-k.

Instead of selecting a fixed number of tokens, it selects enough tokens to reach a cumulative probability threshold.

Suppose the model predicts:

Token	Probability
Paris	60%
Lyon	20%
London	10%
Berlin	5%
Madrid	5%

If:

Top-p = 0.90

The model keeps:

Paris
Lyon
London

because:

0.60 + 0.20 + 0.10 = 0.90

The remaining tokens are discarded.

Top-p adapts dynamically to uncertainty and is widely used in modern language models.

# Context Window

The context window determines how much text the model can see at one time.

Many beginners assume that a model remembers everything.

This is not true.

The model can only directly access the information that fits inside its context window.

Desk Analogy

Imagine a desk.

The larger the desk, the more papers can be placed on it simultaneously.

The smaller the desk, the sooner old papers must be removed.

The context window behaves similarly.

A larger context window allows the model to consider more information at once.

Example

Suppose a model has an 8,000-token context window.

The model can only directly process approximately 8,000 tokens at a time.

As conversations become longer, older information may eventually fall outside the context window.

This is one reason why language models sometimes forget details from earlier parts of long conversations.

# Why Context Windows Matter

Large context windows are important for:

Long conversations
Research papers
Legal documents
Large codebases
PDF analysis
Multi-file projects

A model with a larger context window can reason over much more information simultaneously.

# The Complete Inference Pipeline

When you send a prompt to a model such as Llama or Mixtral, the following steps occur:

The prompt is tokenized.
Tokens are converted into embeddings.
Transformer layers process the embeddings.
The model calculates next-token probabilities.
Temperature, Top-k, and Top-p influence token selection.
A token is chosen.
The new token is appended to the context.
The process repeats until generation is complete.

Every API call follows this general process.

# Connection to Your Internship

When you make your first Groq API call, you will not be training a model.

Instead, you will be performing inference.

Your prompt will be sent to a pretrained model such as:

Llama-3-8B
Mixtral-8x7B

The model will generate responses by repeatedly predicting the next token while applying sampling strategies such as temperature and Top-p.

Understanding these concepts is essential because changing a single parameter can dramatically alter model behavior.

# Key Takeaways

Inference is the process of using a trained model to generate predictions.

Large Language Models generate responses one token at a time.

The model works by repeatedly predicting the most likely next token.

Temperature controls randomness during token selection.

Lower temperatures produce more deterministic outputs.

Higher temperatures produce more creative outputs.

Top-k limits generation to a fixed number of likely tokens.

Top-p limits generation based on cumulative probability.

The context window determines how much information the model can see simultaneously.

Every Groq API call is ultimately a next-token prediction process running inside a transformer.

Humans think in words and ideas.

Computers do not.

A token ID is simply a number.

For example:

```text

tree → 412
```

These IDs are arbitrary.

The model needs a representation that captures meaning.

Embeddings solve this problem.

An embedding is a learned vector representation of a token.

Words that appear in similar contexts tend to receive similar embeddings.

This allows the model to learn relationships such as:

```text
cat ≈ dog

Paris ≈ London

Python ≈ Java
```

Embeddings are the first step toward semantic understanding.

# Why Attention Changed Everything

One of Karpathy's most famous statements is:

> Attention is communication between tokens.

This statement is extremely important.

Before transformers, models relied heavily on memory mechanisms such as RNNs and LSTMs.

These systems struggled with long-range relationships.

Attention introduced a radically different idea.

Instead of trying to remember everything, a token can directly examine other relevant tokens.

Consider the sentence:

```text

```

To understand "it", the model needs information about "animal".

Attention allows the model to focus directly on that relationship.

Every token can exchange information with every other token.

This is why transformers became so powerful.

Every layer allows tokens to exchange information and then process that information.

As layers accumulate, the model develops increasingly sophisticated representations of language.

The deeper the network becomes, the richer these internal representations become.

# Part 11 — System Prompts, Chat Memory, Context Management, and How ChatGPT-Style Assistants Actually Work

## Introduction

At this point in the course, you know how to:

- Send prompts to an LLM
- Use an API
- Choose a model
- Control temperature
- Design better prompts

However, if you build a chatbot using only what we've learned so far, you'll quickly encounter a problem.

Consider the following conversation:

```text
User: My name is Amaima.

Assistant: Nice to meet you, Amaima!

User: What's my name?
```

A human immediately knows the answer.

But if we send only:

```text
What's my name?
```

to the API, the model has no idea.

Why?

Because language models do not automatically remember previous conversations.

This leads us to one of the most important concepts in modern AI systems:

**Context Management.**

---

# The Biggest Beginner Misconception

Many people think:

> ChatGPT remembers everything I say.

This is not actually how LLMs work.

An LLM only sees the text that is included in the current request.

Suppose we send:

```python
messages = [
    {
        "role": "user",
        "content": "What's my name?"
    }
]
```

The model has no information about the earlier message.

Therefore it cannot answer correctly.

---

# Why ChatGPT Appears To Remember

The trick is surprisingly simple.

Instead of sending only the latest message, applications send the entire conversation history.

Example:

```python
messages = [
    {
        "role": "user",
        "content": "My name is Amaima."
    },
    {
        "role": "assistant",
        "content": "Nice to meet you, Amaima!"
    },
    {
        "role": "user",
        "content": "What's my name?"
    }
]
```

Now the model can answer correctly.

The model isn't remembering.

The application is re-sending the history.

This distinction is extremely important.

---

# The Hidden Secret Behind Chatbots

Most beginners imagine:

```text
Conversation
      ↓
Stored Inside Model
```

Reality is closer to:

```text
Conversation History
         ↓
Application
         ↓
Re-sent Every Request
         ↓
Model
```

The application handles memory.

The model processes context.

---

# What Is Context?

Context is all information visible to the model during a request.

This may include:

- System prompts
- Previous messages
- User messages
- Retrieved documents
- Tool outputs

Everything inside the context window influences generation.

Everything outside the context window effectively does not exist.

# The Context Window Problem

Unfortunately, memory is not infinite.

Remember context windows?

Suppose a model supports:

```text

```

Eventually the conversation becomes longer than 8,000 tokens.

Now we have a problem.

The entire conversation no longer fits.

# What Happens When Context Fills Up?

Imagine a desk that can only hold ten papers.

Once paper eleven arrives:

Something must be removed.

Language models behave similarly.

When the context window becomes full:

Older information must eventually be discarded.

---

# Why Models Sometimes "Forget"

Users often say:

> The model forgot what I told it earlier.

Usually this happens because:

- The conversation became too long.
- Earlier messages fell outside the context window.

The model isn't forgetting in a human sense.

The information is simply no longer visible.

---

# Memory Management Strategies

Professional AI systems use several techniques to handle long conversations.

---

## Strategy 1: Truncation

Remove old messages.

Example:

```text
Keep newest messages.
Delete oldest messages.
```

Simple but potentially loses important information.

---

## Strategy 2: Summarization

Instead of storing every message:

Create a summary.

Example:

```text
User is learning calculus.

User struggles with integration.

Previous topics:
- Derivatives
- Limits
- Chain Rule
```

A short summary preserves important information.

---

## Strategy 3: Long-Term Memory

Store important facts separately.

Example:

```text
User Name = Amaima

Interest = AI

Current Topic = Calculus
```

Later, these facts can be reinserted into context.

Many advanced assistants use this approach.
