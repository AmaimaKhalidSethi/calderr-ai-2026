# The Core Misconception

Many beginners think:

> "ChatGPT is already an AI Agent."

This is not actually true.

ChatGPT is primarily a language model.

An agent is something much more powerful.

The confusion exists because modern chatbots *feel* intelligent.

They can:

- Answer questions
- Write code
- Explain concepts
- Generate plans
- Hold conversations

This makes it seem like they are actively thinking.

But internally something very different is happening.

---

# First Principles: What Is An Agent?

Before AI existed, the word "agent" already existed in computer science.

An agent is:

> A system that can perceive its environment, make decisions, and take actions to achieve a goal.

Notice that language is not part of this definition.

The essential pieces are:

1. Observe
2. Decide
3. Act

Let's use a human example.

---

## Human Agent Example

Imagine you want coffee.

You:

### Perceive

You realize:

> "I'm feeling sleepy."

### Plan

You think:

> "Coffee might help."

### Act

You walk to the kitchen.

### Observe

You discover:

> "No coffee beans."

### Re-plan

You decide:

> "I'll order coffee online."

### Act Again

You place an order.

### Observe Again

Coffee arrives.

Goal achieved.

---

This process is called:

```text
Observe
   ↓
Plan
   ↓
Act
   ↓
Observe
   ↓
Plan Again
   ↓
Act Again
```

The loop continues until the goal is reached.

This is agent behavior.

---

# What Is A Language Model?

Now compare that with an LLM.

An LLM does something much simpler.

Input:

```text
User Prompt
```

Output:

```text
Generated Text
```

Diagram:

```text
Prompt
   ↓
LLM
   ↓
Text
```

That's it.

No actions.

No goals.

No environment.

No planning loop.

No memory of the real world.

No observation after acting.

---

# Why ChatGPT Is Not Inherently An Agent

Suppose you ask:

```text
Book me the cheapest flight to Dubai.
```

A pure language model cannot actually:

- Open websites
- Compare prices
- Login to accounts
- Fill forms
- Make purchases

It can only generate text like:

```text
You could check Expedia or Skyscanner.
```

The model is describing actions.

It is not performing actions.

This distinction is extremely important.

---

# The "Next Token Machine"

Internally, an LLM is solving a surprisingly simple problem.

It predicts:

> What token comes next?

Example:

Input:

```text
The capital of France is
```

The model predicts:

```text
Paris
```

Then:

```text
Paris.
```

Then:

```text
Paris. It
```

Then:

```text
Paris. It is
```

And so on.

The model repeatedly predicts the next token.

Diagram:

```text
Token 1
   ↓
Token 2
   ↓
Token 3
   ↓
Token 4
   ↓
...
```

Everything that feels like intelligence emerges from this process.

But notice something.

The model is not:

- Checking reality
- Taking actions
- Running experiments
- Gathering evidence

It is generating probable text.

---

# Limitation #1: No Real-Time Information

Suppose an LLM was trained in 2025.

You ask:

```text
Who won yesterday's cricket match?
```

The model may not know.

Why?

Because training already ended.

The model cannot magically access new information.

Without tools:

```text
Question
   ↓
LLM
   ↓
Guess
```

This creates hallucinations.

---

# Limitation #2: No External Actions

Ask:

```text
Send an email to my professor.
```

The LLM can generate:

```text
Dear Professor...
```

But it cannot actually send the email.

Why?

Because it has no connection to:

- Gmail
- Outlook
- SMTP servers

It only generates text.

---

# Limitation #3: No Verification

Suppose you ask:

```text
What is the current Bitcoin price?
```

Without internet access:

The model must rely on training knowledge.

It cannot verify whether the answer is correct.

Humans can.

Agents can.

Pure LLMs cannot.

---

# Limitation #4: No Long-Term Goal Pursuit

Imagine:

```text
Find an internship.
```

This goal may require:

1. Searching websites
2. Comparing jobs
3. Reading requirements
4. Updating resume
5. Sending applications
6. Tracking responses

A pure chatbot cannot continuously pursue that objective.

It responds once.

Conversation ends.

Agent systems can continue working toward the goal.

---

# Why Prompt Engineering Was Not Enough

Around 2022–2023, researchers discovered something interesting.

By writing better prompts, LLMs became dramatically better.

Example:

Instead of:

```text
Solve this problem.
```

People wrote:

```text
Think step by step.
```

Performance improved significantly.

This led to:

- Chain-of-Thought Prompting
- Few-Shot Prompting
- Reasoning Prompts

Researchers hoped prompting alone might solve many problems.

But eventually they hit a wall.

---

# Example: Mathematical Research Problem

Prompt:

```text
Find the answer.
Think step by step.
```

The model can reason.

But what if it needs:

- A calculator
- A database
- Internet access
- A search engine

Reasoning alone is insufficient.

The model needs tools.

---

# The Emergence Of Tool Use

Researchers started asking:

> What if the model could call external tools?

Imagine:

```text
User:
What's the weather today?
```

Instead of guessing:

```text
LLM
   ↓
Weather API
   ↓
Real Data
   ↓
Answer
```

Now the answer becomes grounded in reality.

This was a huge breakthrough.

---

# Tool-Augmented AI

The architecture changed from:

```text
User
   ↓
LLM
   ↓
Response
```

to:

```text
User
   ↓
LLM
   ↓
Tool
   ↓
Result
   ↓
LLM
   ↓
Response
```

The model could now:

- Search the web
- Run Python
- Query databases
- Read files
- Use APIs
- Execute code

Suddenly LLMs became much more useful.

---

# But Another Problem Appeared

How does the model know:

- WHEN to use a tool?
- WHICH tool to use?
- WHAT to do after the tool returns results?

Example:

```text
Question:
What is the weather in Lahore and what clothes should I wear?
```

The model must:

1. Realize weather information is required.
2. Call weather API.
3. Read result.
4. Reason about clothing.
5. Produce final answer.

This requires decision-making.

Decision-making requires an agent loop.

---

# Historical Motivation Behind ReAct

Researchers noticed two different strengths of LLMs.

### Strength 1: Reasoning

The model can think.

Example:

```text
If A > B
and B > C
then A > C
```

Reasoning works.

---

### Strength 2: Acting

The model can invoke tools.

Example:

```text
Search()
Calculator()
DatabaseQuery()
```

But early systems separated these abilities.

Reasoning and acting were disconnected.

This caused poor performance.

---

# The Key Insight

Humans do not:

```text
Think for 10 minutes
then act once
```

Instead we:

```text
Think
Act
Observe
Think Again
Act Again
Observe Again
```

This loop is crucial.

Researchers realized AI should behave similarly.

This insight led directly to the ReAct framework.

---

# What ReAct Introduced

ReAct stands for:

```text
Reason + Act
```

The core idea:

```text
Thought
   ↓
Action
   ↓
Observation
   ↓
Thought
   ↓
Action
   ↓
Observation
```

Instead of generating one final answer, the model repeatedly reasons and acts.

This became the foundation of modern AI agents.

# Interview-Level Understanding

A common interview question:

### What is the difference between an LLM and an AI Agent?

Strong answer:

> An LLM is a language generation system that predicts tokens based on context. An AI Agent is a system that uses an LLM (or another decision-making component) to perceive information, plan actions, interact with tools or environments, observe outcomes, and iteratively work toward a goal.

# Introduction

In Part 1, we learned a critical idea:

> An LLM is not automatically an agent.

An LLM can generate text.

An agent can pursue goals.

The missing piece is a continuous decision-making cycle.

This cycle is called:

```text
Agent Loop
```

Every modern AI agent—from ChatGPT with tools to autonomous coding assistants—is built around some variation of this loop.

Before learning frameworks like LangChain, CrewAI, AutoGen, or OpenAI Agents, you must understand this loop deeply.

Because every framework is simply an abstraction over the same underlying process.

---

# Why Does An Agent Need A Loop?

Imagine a robot whose job is:

```text
Bring me a bottle of water.
```

Could the robot simply think once and succeed?

No.

Let's see why.

---

## Scenario

The robot thinks:

```text
The water is probably in the kitchen.
```

It goes there.

But then discovers:

```text
No water bottle.
```

Now what?

If the robot cannot observe and re-plan:

```text
Mission Failed
```

The robot needs to:

1. Observe reality
2. Update its understanding
3. Make a new plan
4. Act again

This cycle repeats until the goal is achieved.

---

# The Fundamental Pattern Of Intelligence

Interestingly, humans operate exactly the same way.

Consider:

```text
Goal:
Pass an AI Internship
```

You don't instantly know everything.

Instead:

```text
Study
 ↓
Observe confusion
 ↓
Ask questions
 ↓
Learn
 ↓
Practice
 ↓
Observe mistakes
 ↓
Improve
```

This process is recursive.

Intelligence is often less about knowing everything upfront and more about continuously adapting.

That adaptation is what the Agent Loop enables.

---

# The Four Core Stages

The standard agent loop contains four phases:

```text
Perceive
   ↓
Plan
   ↓
Act
   ↓
Observe
   ↓
Repeat
```

Let's study each one separately.

---

# Phase 1 — Perceive

---

Perception means:

> Gathering information about the current state of the environment.

Humans constantly perceive.

Examples:

```text

```

These observations influence your decisions.

Agents must do the same.

---

## AI Example

User asks:

```text

```

Before planning anything, the agent must perceive:

```text
What city?
What date?
What weather data exists?
```

Without information:

```text
No meaningful decision can be made.
```

---

## Human Analogy

Imagine driving a car blindfolded.

You can still make decisions.

But they will be terrible.

Why?

Because planning without perception is guessing.

Agents need perception to ground themselves in reality.

# Phase 2 — Plan

---

## What Is Planning?

Planning means:

> Deciding what should happen next.

After gathering information, the agent must determine:

```text
What action best moves me toward the goal?
```

This is where the LLM usually becomes the brain of the system.

---

## Example

User asks:

```text
Find the cheapest flight to Dubai.
```

The agent perceives:

```text
Origin: Islamabad
Destination: Dubai
Travel Date: July
```

Now it plans:

```text
1. Search flights
2. Compare prices
3. Rank options
4. Return recommendation
```

Notice:

The agent still hasn't acted.

It is only deciding.

---

# Why Planning Matters

Without planning:

```text
Input
 ↓
Random Action
```

This produces chaotic behavior.

Planning creates intentional behavior.

---

## Human Analogy

Suppose your goal is:

```text
Score A+ in Calculus
```

You don't randomly:

```text
Watch random videos
Read random chapters
Solve random questions
```

You make a plan:

```text
Understand theory
Practice examples
Solve past papers
Review mistakes
```

Agents work similarly.

---

# Planning Is Prediction

Internally, planning is often:

```text
"If I do X,
what is likely to happen?"
```

The LLM uses patterns learned during training to estimate useful next steps.

# The Problem Before ReAct

To appreciate ReAct, we must understand what came before it.

Around 2022, researchers discovered something surprising.

LLMs became dramatically better when asked to explain their reasoning.

Example:

Instead of asking:

```text
What is 27 × 43?
```

Researchers asked:

```text
What is 27 × 43?

Think step by step.
```

Performance improved.

This technique became known as:

```text
Chain of Thought (CoT)
```

---

# What Is Chain of Thought?

Chain of Thought encourages the model to reveal intermediate reasoning.

Example:

```text
Question:
A store sells 3 apples for $6.
How much do 9 apples cost?
```

Without reasoning:

```text
$18
```

With reasoning:

```text
3 apples cost $6

1 apple costs $2

9 apples cost $18
```

The answer is identical.

But reasoning improves accuracy.

---

# Why Did Chain of Thought Work?

Because complex problems can often be broken into smaller problems.

Humans do the same thing.

Imagine calculating:

```text
847 × 193
```

You don't instantly know the answer.

You perform intermediate steps.

Chain of Thought lets the model simulate those steps.

---

# The Limitation Of Chain of Thought

However, researchers soon encountered a serious problem.

Reasoning alone is not enough.

---

## Example 1

Question:

```text
Who won yesterday's cricket match?
```

The model can reason perfectly.

But reasoning cannot create missing information.

The model still needs:

```text
Current data
```

---

## Example 2

Question:

```text
What is the current Bitcoin price?
```

Again:

```text
Reasoning ≠ Real-Time Knowledge
```

The model needs external information.

---

## Example 3

Question:

```text
Find the cheapest hotel in Islamabad this weekend.
```

The model needs:

- Search
- Comparison
- Live pricing

Reasoning alone cannot solve the task.

---

# The Opposite Problem

Researchers then experimented with giving models tools.

Example:

```text
Search()
Calculator()
Wikipedia()
```

Now the model could access information.

Great.

Problem solved?

Not quite.

---

# Why Tool Use Alone Failed

Imagine a student taking an exam.

Would giving them:

```text
Calculator
Internet
Books
```

````python
# ============================================
# DAY 2 - PART 1
# Rule-Based ReAct Agent
# ============================================

def calculator_tool(expression):
    print("[TOOL] Calculator Tool Called")
    return eval(expression)


def weather_tool(city):
    print("[TOOL] Weather Tool Called")
    return f"The weather in {city} is Sunny, 35°C"


def search_tool(query):
    print("[TOOL] Search Tool Called")
    return f"Search results for: {query}"


def choose_tool(user_input):

    print("\n[THOUGHT]")
    print("Analyzing user input...")

    if any(char.isdigit() for char in user_input):
        print("Detected numbers -> Calculator Tool")
        return "calculator"

    elif "weather" in user_input.lower():
        print("Detected weather query -> Weather Tool")
        return "weather"

    else:
        print("Defaulting to Search Tool")
        return "search"


def execute_tool(tool_name, user_input):

    print("\n[ACTION]")
    print(f"Executing: {tool_name}")

    if tool_name == "calculator":
        return calculator_tool(user_input)

    elif tool_name == "weather":

        city = (
            user_input
            .lower()
            .replace("weather in", "")
            .strip()
            .title()
        )

        return weather_tool(city)

    elif tool_name == "search":
        return search_tool(user_input)


def agent(user_input):

    print("\n" + "=" * 50)
    print("NEW AGENT RUN")
    print("=" * 50)

    print(f"\n[USER INPUT]\n{user_input}")

    tool = choose_tool(user_input)

    result = execute_tool(tool, user_input)

    print("\n[OBSERVATION]")
    print(result)

    print("\n[FINAL ANSWER]")
    print(result)


# TESTS
agent("25 * 12")
agent("weather in lahore")
agent("latest ai news")
````

````python
# ============================================
# DAY 2 - PART 2
# LLM-Based Tool Selection (Groq)
# ============================================

import os

from groq import Groq

client = Groq(api_key=os.getenv("GROQ_API_KEY"))


def calculator_tool(expression):
    print("[TOOL] Calculator Tool Called")
    return str(eval(expression))


def weather_tool(city):
    print("[TOOL] Weather Tool Called")
    return f"The weather in {city} is Sunny, 35°C"


def search_tool(query):
    print("[TOOL] Search Tool Called")
    return f"Search results for: {query}"


def choose_tool_llm(user_input):

    prompt = f"""
You are an AI Agent.

Available Tools:

calculator
- mathematical calculations

weather
- weather information

search
- general information retrieval

User Question:
{user_input}

Return ONLY ONE WORD:

calculator
weather
search
"""

    print("\n[THOUGHT]")
    print("Sending reasoning request to LLM...")

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    tool = (
        response
        .choices[0]
        .message.content
        .strip()
        .lower()
    )

    print(f"LLM Selected Tool -> {tool}")

    return tool


def execute_tool(tool_name, user_input):

    print("\n[ACTION]")
    print(f"Executing {tool_name}")

    if tool_name == "calculator":
        return calculator_tool(user_input)

    elif tool_name == "weather":

        city = (
            user_input
            .replace("weather in", "")
            .strip()
        )

        return weather_tool(city)

    else:
        return search_tool(user_input)


def agent(user_input):

    print("\n" + "=" * 50)
    print("LLM AGENT RUN")
    print("=" * 50)

    print(f"\n[USER INPUT]\n{user_input}")

    tool = choose_tool_llm(user_input)

    result = execute_tool(tool, user_input)

    print("\n[OBSERVATION]")
    print(result)

    print("\n[FINAL ANSWER]")
    print(result)


# TEST
# agent("Will I need an umbrella today?")
agent("587 * 234")
agent("latest ai news")
````

````python
from groq import Groq

# ==========================================
# CONFIG
# ==========================================

client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

# ==========================================
# TOOLS
# ==========================================

def calculator_tool(expression):

    print("\n[TOOL] Calculator")

    try:
        return str(eval(expression))

    except Exception as e:
        return f"Calculation Error: {e}"


def weather_tool(city):

    print("\n[TOOL] Weather")

    return f"Weather in {city}: Sunny, 35°C"


def search_tool(query):

    print("\n[TOOL] Search")

    return f"Search Result for: {query}"


TOOLS = {
    "calculator": calculator_tool,
    "weather": weather_tool,
    "search": search_tool
}

# ==========================================
# TOOL SELECTION
# ==========================================

def choose_tool(user_input):

    prompt = f"""
You are an AI agent.

Available tools:

calculator
weather
search

User Query:
{user_input}

Return ONLY ONE WORD:

calculator
weather
search

No explanation.
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    tool = (
        response
        .choices[0]
        .message.content
        .strip()
        .lower()
    )

    return tool

# ==========================================
# FINAL ANSWER GENERATION
# ==========================================

def generate_answer(question, observation):

    prompt = f"""
Question:
{question}

Tool Output:
{observation}

Provide a final answer.
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response.choices[0].message.content

# ==========================================
# EXECUTE TOOL
# ==========================================

def execute_tool(tool_name, user_input):

    if tool_name == "calculator":
        return calculator_tool(user_input)

    elif tool_name == "weather":

        city = (
            user_input
            .lower()
            .replace("weather in", "")
            .strip()
            .title()
        )

        return weather_tool(city)

    elif tool_name == "search":
        return search_tool(user_input)

    else:
        return "Unknown Tool"

# ==========================================
# AGENT
# ==========================================

def react_agent(user_input):

    print("\n" + "="*60)
    print("AGENT STARTED")
    print("="*60)

    print("\n[USER INPUT]")
    print(user_input)

    # -----------------------
    # THOUGHT
    # -----------------------

    print("\n[THOUGHT]")
    print("Selecting tool...")

    tool_name = choose_tool(user_input)

    print(f"Chosen Tool -> {tool_name}")

    # -----------------------
    # ACTION
    # -----------------------

    print("\n[ACTION]")

    observation = execute_tool(
        tool_name,
        user_input
    )

    # -----------------------
    # OBSERVATION
    # -----------------------

    print("\n[OBSERVATION]")
    print(observation)

    # -----------------------
    # FINAL ANSWER
    # -----------------------

    answer = generate_answer(
        user_input,
        observation
    )

    print("\n[FINAL ANSWER]")
    print(answer)

    return answer


# ==========================================
# TESTS
# ==========================================

react_agent("50 * 12")

react_agent("weather in lahore")

react_agent("latest ai news")
````
